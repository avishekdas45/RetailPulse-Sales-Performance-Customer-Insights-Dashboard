```python
cursor.execute('''SELECT * FROM store;''')
rows = cursor.fetchall()
df = pd.DataFrame
column = (desc[0] for desc in cursor.description)
df(rows, columns=column)
```

### 1. What is the total sales and profit generated by the company?

```python
cursor.execute('''
    SELECT SUM(sales) as total_sales,
    SUM(profit) AS total_profit
    FROM store;
    ''')

rows = cursor.fetchall()
column = (desc[0] for desc in cursor.description)
df(rows, columns= column).round(2)
```

### 2. Which are the top 5 products with the highest sales?

```python
cursor.execute('''
    SELECT product_name, ROUND(SUM(sales),2) as total_sales
    FROM store
    GROUP BY product_name
    ORDER BY total_sales DESC
    LIMIT 05;
    ''')

rows = cursor.fetchall()
column = (desc[0] for desc in cursor.description)
df(rows,columns= column)
```

### 3. What is the total sales by region and state?

```python
cursor.execute('''
    SELECT region, state, ROUND(SUM(sales),2) AS total_sales
    FROM store
    GROUP BY region, state
    ORDER BY region, state;
    ''')

rows = cursor.fetchall()
column = (desc[0] for desc in cursor.description)
df(rows, columns= column)
```

### 4. Which are the 5 least profitable sub-categories?

```python
cursor.execute('''
    SELECT Sub_Category, ROUND(SUM(profit),2) AS total_profit
    FROM store
    GROUP BY Sub_Category
    ORDER BY total_profit DESC
    LIMIT 05;
    ''')

rows = cursor.fetchall()
column = (desc[0] for desc in cursor.description)
df(rows, columns= column)
```

### 5. What is the average discount given for each category?

```python
cursor.execute('''
    SELECT category, AVG(discount) AS average_discount
    FROM store
    GROUP BY category
    ''')
rows = cursor.fetchall()
column = (desc[0] for desc in cursor.description)
df(rows, columns = column)
```

### 6. Who are the top 5 customers based on profit?

```python
cursor.execute('''
    SELECT Customer_Name, SUM(profit) AS total_profit
    FROM store
    GROUP BY Customer_Name
    ORDER BY total_profit DESC
    LIMIT 05
    ''')

rows = cursor.fetchall()
column =  (desc[0] for desc in cursor.description)
df(rows, columns =  column)
```

### 7. How much sales and quantity has each sub-category generated?

```python
cursor.execute('''
    SELECT Sub_Category, SUM(Quantity) AS total_quantity, SUM(sales) AS total_sales
    FROM store
    GROUP BY Sub_Category
    ORDER BY total_sales DESC, total_quantity desc
    ''')

rows = cursor.fetchall()
column =  (desc[0] for desc in cursor.description)
df(rows, columns =  column)
```

### 9. How many orders were placed using each shipping mode?

```python
cursor.execute('''
    SELECT Ship_Mode, COUNT(Ship_Mode) as Total_Transaction
    FROM store
    GROUP BY Ship_Mode
    ''')

rows = cursor.fetchall()
column =  (desc[0] for desc in cursor.description)
df(rows, columns =  column)
```

### 10. Which 10 cities are the most profitable?

```python
cursor.execute('''
    SELECT city, ROUND(SUM(profit),2) AS total_profit
    FROM store
    GROUP BY city
    ORDER BY total_profit DESC
    LIMIT 10
    ''')

rows = cursor.fetchall()
column =  (desc[0] for desc in cursor.description)
df(rows, columns =  column)
```

### 11. Which orders resulted in a loss (negative profit)?

```python
cursor.execute('''
    SELECT order_id, product_name, profit
    FROM store
    WHERE profit < 0
    ''')

rows = cursor.fetchall()
column =  (desc[0] for desc in cursor.description)
df(rows, columns =  column)
```
